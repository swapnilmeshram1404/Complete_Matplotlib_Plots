```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)

b = a ** 4
```

```python
x = np.arange(0,10)

y = 2 * x
```

```python
fig , axes = plt.subplots(nrows = 1 , ncols = 2)  #empty subplots
```

```python
fig , axes = plt.subplots(nrows = 1 , ncols = 2)

axes[0].plot(x,y)    #ploting x-y relationship graph at index 0 position
```

```python
fig , axes = plt.subplots(nrows = 1 , ncols = 2)

axes[1].plot(a,b)    #ploting a-b relationship graph at index 1 position
```

```python
fig , axes = plt.subplots(nrows = 1 , ncols = 2)

axes[0].plot(x,y)
                    #ploting both subplots together
axes[1].plot(a,b)
```

```python
#Two dimensional figure 
```

```python
fig, axes = plt.subplots( nrows = 2 , ncols = 2 )

axes [0][0].plot(x,y)
```

```python
fig, axes = plt.subplots( nrows = 2 , ncols = 2 )

axes [0][0].plot(x,y)

axes[1][1].plot(a,b)
```

```python
#again one dimensional figure
```

```python
fig , axes = plt.subplots( nrows =3 , ncols = 1 )

axes[0].plot(x,y)

axes[2].plot(a,b)
```

```python
fig , axes = plt.subplots(nrows = 2 , ncols = 2)

axes[0][0].plot(x,y)

axes[0][1].plot(a,b)

axes[1][0].plot(a,b)

axes[1][1].plot(x,y)

fig.tight_layout()          #for good spacing between subplots
```

```python
fig , axes = plt.subplots(nrows = 2 , ncols = 2)

axes[0][0].plot(x,y)

axes[0][1].plot(a,b)

axes[1][0].plot(a,b)

axes[1][1].plot(x,y)

fig.subplots_adjust(wspace = 0.5, hspace = 0.5); #for manually adjusting spacing between subplots 
```

```python
#setting titles and x,y axis labels to all subplots
```

```python
fig , axes = plt.subplots(nrows = 2 , ncols = 2, dpi = 150)

axes[0][0].plot(x,y)
axes[0][0].set_xlabel("1st plot x axis")
axes[0][0].set_ylabel("1st plot y axis")
axes[0][0].set_title("INDEX 0,0 PLOT")

axes[0][1].plot(a,b)
axes[0][1].set_xlabel("2nd plot x axis")
axes[0][1].set_ylabel("2nd plot y axis")
axes[0][1].set_title("INDEX 0,1 PLOT")

axes[1][0].plot(a,b)
axes[1][0].set_xlabel("3rd plot x axis")
axes[1][0].set_ylabel("3rd plot y axis")
axes[1][0].set_title("INDEX 1,0 PLOT")

axes[1][1].plot(x,y)
axes[1][1].set_xlabel("4th plot x axis")
axes[1][1].set_ylabel("4th plot y axis")
axes[1][1].set_title("INDEX 1,1 PLOT")

fig.subplots_adjust(wspace = 0.8 , hspace = 0.9);
```
