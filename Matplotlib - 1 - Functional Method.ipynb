```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)

y = 2 * x


print(x)

print(y)
```

# Basic Plot

```python
plt.plot(x,y)

plt.show()
```

# Adding Title and X , Y Axis Lables

```python
plt.plot(x,y)

plt.title("Some Title")

plt.xlabel("X Axis")

plt.ylabel("Y Axis");
```

# Adjusting values on X and Y Axis and Saving The Plot

```python
plt.plot(x,y)

plt.title("Some Title")

plt.xlabel("X Axis")

plt.ylabel("Y Axis")

plt.xlim(0 , 10)

plt.ylim(0,20)

plt.savefig("myfirstplt.png");
```
