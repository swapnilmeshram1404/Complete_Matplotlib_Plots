```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure(dpi=150)

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x, color = "#9D41EE", label = "x vs x") #RGB HEX CODE

axes.plot(x, x+1, color ="#EE6F41" , label = "x vs x+1")

plt.legend(loc = (1,1))
```

```python
fig = plt.figure(dpi=150)

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x, color = "#9D41EE", linewidth = 2, linestyle = "-." , marker = "o") 

axes.plot(x, x+1, color ="#EE6F41" , linewidth = 2, linestyle = "--" , marker = "+", markersize = 10);

```

```python
fig = plt.figure(dpi=150)

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x, color = "#9D41EE", linewidth = 2, linestyle = "-." , marker = "o", 
          markerfacecolor = "green", markeredgewidth = 2 , markeredgecolor = "orange") 
```

```python
fig = plt.figure(dpi=150)

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x, color = "#9D41EE", linewidth = 2, linestyle = "-." , marker = "s", #for square
          markerfacecolor = "green", markeredgewidth = 2 , markeredgecolor = "orange") 
```
