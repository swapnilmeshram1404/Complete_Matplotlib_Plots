```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)

b = a ** 4
```

```python
x = np.arange(0,10)

y = 2 * x
```

```python
fig = plt.figure() #gives blur image

axes = fig.add_axes([0,0,1,1])

axes.plot(a,b);
```

```python
fig = plt.figure(dpi = 150) #gives sharp image

axes = fig.add_axes([0,0,1,1])

axes.plot(a,b);
```

```python
fig = plt.figure(figsize = (8,3) , dpi =150) #gives horizontal image

axes = fig.add_axes([0,0,1,1])

axes.plot(a,b);
```

```python
fig = plt.figure(figsize = (1,3) , dpi =150) #gives vertical image

axes = fig.add_axes([0,0,1,1])

axes.plot(a,b);
```

```python
fig = plt.figure(figsize = (2,2) , dpi =150) 

axes = fig.add_axes([0,0,1,1])

axes.plot(a,b)

fig.savefig("mysecondfigure.png" , bbox_inches = "tight") #saving figure that looks good from all the above figures
```
