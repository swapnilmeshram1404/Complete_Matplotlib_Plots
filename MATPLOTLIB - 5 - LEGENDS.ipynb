```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])


axes.plot(x,x)

axes.plot(x , x**2)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])


axes.plot(x,x, label = "x vs x")

axes.plot(x , x**2, label = "x vs x**2")

axes.legend()
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])


axes.plot(x,x, label = "x vs x")

axes.plot(x , x**2, label = "x vs x**2")

axes.legend(loc = (1,1))
```
