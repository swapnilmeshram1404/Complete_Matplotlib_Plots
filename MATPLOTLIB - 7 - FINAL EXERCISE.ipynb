### PLOTING e = MC**2

```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
m = np.linspace(0,10,11)
```

```python
c = 3 * 10**8
```

```python
e = m * c**2
```

```python
e
```

```python
fig = plt.figure(dpi = 150)

axes = fig.add_axes([0,0,1,1])

axes.plot(m,e, lw = 2, color = "purple" , marker = "s" )

axes.set_title("E = MC**2")

axes.set_xlabel("mass in grams")

axes.set_ylabel("energy in jules");
```

### Q2 : US BOND YEILD CURVE PLOTTING

```python
labels = ["1M" , "3M" , "6M", "1YR", "2YR", "3YR", "5YR", "7YR", "10YR", "20YR", "30YR"]

july_16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89,4.89, 4.95, 4.99, 5.05, 5.21, 5.14]

july_16_2020 = [0.12, 0.11, 0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

```python
fig = plt.figure(dpi = 100)

axes = fig.add_axes([0,0,1,1,])

axes.plot(labels , july_16_2007 , lw = 3 , color = "red", marker = "x" , markersize = 15, label = "july_16_2007")

axes.plot(labels , july_16_2020 , lw = 3 , color = "blue", marker = "s" , markersize = 8 , label = "july_16_2020")

plt.legend(loc = (1.05,0.5));
```

```python
fig , axes = plt.subplots(nrows = 2 , ncols = 1 , dpi = 100, figsize = (15,10))

axes[0].plot(labels , july_16_2007 , lw = 3 , color = "red", marker = "x" , markersize = 15)

axes[0].set_title("july_16_2007")
axes[0].set_xlabel("year")
axes[0].set_ylabel("returns")


axes[1].plot(labels , july_16_2020 , lw = 3 , color = "blue", marker = "s" , markersize = 8)

axes[1].set_title("july_16_2020")
axes[1].set_xlabel("year")
axes[1].set_ylabel("returns")

plt.subplots_adjust( hspace = 0.3)
```
